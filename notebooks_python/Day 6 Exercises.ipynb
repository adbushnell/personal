```python
#elementwise ops q1
df_monster['AverageStat'] = df_monster.apply(lambda x: x.iloc[1:3].mean(), axis=1)
df_monster
```

```python
df_monster['MinStat'] = df_monster.apply(lambda x: x.iloc[1:3].min(), axis=1)
df_monster
```

```python
df_monster['MaxStat'] = df_monster.apply(lambda x: x.iloc[1:3].max(), axis=1)
df_monster
```

```python
#aggregation q1
#you would filter the average stat to find the maximum
df_monster.loc[df_monster['AverageStat']==df_monster['AverageStat'].max()]
```
