## Interfaces
---
There are two interfaces:
- MATLAB style (stateful interface)
- Object oriented (nonstateful interface): needs to be mor explicit

Other than the plots in this section we will use the 2nd one

```python
#MATLAB interface
plt.figure() #set the canvas up
plt.subplot(2,1,1) #2 rows one column, start at 1
# we have two plots then (one on each row)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2) #now second row
plt.plot(x,np.cos(x))
#its kinda ugly though
```

```python
#OO Interface
fig, axarr = plt.subplots(2,1) #figure and axis (axisarray = axarr)
axarr[0].plot(x,np.sin(x))
axarr[1].plot(x,np.cos(x))
#probably prettier code
```

## Line plots
---
Plot a line through the x and y values, we have some keywords that we can change too:
- color
- alpha (transparency)
- linestyle
- marker (points)
- linewidth
- label


```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x,np.sin(x), alpha=0.5)
ax.plot(x,np.cos(x), color='red')
ax.plot(x,np.sin(x+2), linestyle='--')
ax.plot(x,np.sin(x+3), marker='o')
ax.plot(x,np.sin(x+4), label='madness')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, np.sin(x), alpha=0.25, color='purple', marker='^', linewidth=10, label='Wavy')
ax.legend()
```

## Set x and y limits

```python
def create_base_plot():
    x=np.linspace(0,10,100)
    y=np.sin(x)
    fig, ax = plt.subplots(figsize=(12,6))
    ax.plot(x,y)
    return fig, ax
```

```python
fig, ax = create_base_plot()
ax.set_xlim(0,2)
ax.set_ylim(-0.5,0.2)

#x.set_xlim(left=1,right=3)
#x.set_ylim(ymin=-0.5,ymax=2)
```

```python
fig, ax = create_base_plot()
ax.set_xlabel("X Label")
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

## Histograms

```python
x = np.random.normal(0,5,2000)
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(x,bins=100)
```

```python
# Plotting the histogram (density) against the normal distribution
data = np.random.normal(0, 5, 2000) # 2000 samples with mean 0, std 5
x = np.linspace(-15,15,100)
y = ss.norm.pdf(x, loc=0, scale=5)

 

fig, ax = plt.subplots(figsize=(12,6))
ax.hist(data, bins=30, density=True)
ax.plot(x,y) # pdf
```

### 2-D histograms

```python
mean = [0,0]
cov = [[1,1],[1,2]]
data = np.random.multivariate_normal(mean, cov, 1000000)
```

```python
x = data[:,0]
y = data[:,1]

fig, ax = plt.subplots(figsize=(12,6))
hist_data = ax.hist2d(x,y, bins =500)
plt.colorbar(hist_data[3], ax=ax)
```

## Contour plots

```python
def f(x,y):
    z = 2*x**2 + y**2
    return z
```

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)

X, Y = np.meshgrid(x,y)
```

```python
Z = f(X,Y)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
contour_set = ax.contour(X,Y,Z, levels=10)
ax.clabel(contour_set)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
contour_set = ax.contourf(X,Y,Z, levels=10, cmap='magma') #looks like looney tunes 
```

#### Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,6))
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
```

Concept Check
-------------
- When would you use a line chart vs a scatter plot?
- When would you use a bar chart vs a box plot?
- When would you use a histogram vs a box plot?
- When would you use a stacked bar chart vs a stacked area plot?

```python
#line is a single variable over some index like time, scatter is for paired data
#bar chart is good for count data or cumulative data, box plot good at analysing distributions
#histogram when you have realisations with an index that can be split into bins, box plots have pre defined categories
#stacked bars are good for seeing the cumulative effect of an index's categories, stacked area shows the cumulative effect of all indexes over time 
```
